```python
import time
import random as rd
```

```python
def attack(power,name):
    attack_choice = rd.randint(1, 5)
        
    if attack_choice == 1:
        power = 10
        name = 'Kick'
    if attack_choice == 2:
        power = 5
        name = 'Puch'
    if attack_choice == 3:
        power = 7
        name = 'Double Punch'
    if attack_choice == 4:
        power = 12
        name = 'Double Kick'
    if attack_choice == 5:
        power = -2
        name = 'Stole HP and leach'
        
    return power,name
```

```python
def fight_start(player1,player2):
    
    # Few var setup
    health1 = 100
    health2 = 100
    p1_c = 0
    p2_c = 0
    attack_power = 0
    attack_name = ''
    
    print('-----------LET THE BATTLE BEGIN-----------')
    time.sleep(2)
    
    # main game loop
    while True:
        
        
        
        # chooses on random which player attacks 
        player_choice = rd.randint(1, 2)
        
        attack_power,attack_name = attack(attack_power,attack_name)

        # Player attacking
        if player_choice == 1:
            health2 -= attack_power
            p1_c += 1
            print(f'----{player1} {attack_name}ed {player2}----')
            print(f'Status : {player1} = {health1} HP and {player2} = {health2} HP')

        if player_choice == 2:
            health1 -= attack_power
            p2_c += 1
            print(f'----{player2} {attack_name}ed {player1}----')
            print(f'Status : {player1} = {health1} HP and {player2} = {health2} HP')
        
        # Just a blank line and time break      
        print('')
        time.sleep(1.25)
        
        # Break the loop
        if ((health1 <= 0) or (health2 <= 0)):
            break
        
        
    
    # Game Results
    if health1 == 0:
        print(f'----{player2} Won----')
        print(f'{player2} attacked {p2_c} times')
        print(f'{player1} is Dead')
        print(f'{player1} attacked {p1_c} times')
    else:
        print(f'----{player1} Won----')
        print(f'{player1} attacked {p1_c} times')
        print(f'{player2} is Dead')
        print(f'{player2} attacked {p2_c} times')
        
    
```

```python
# Main game area


# Getting player 1 and player 2 name
# player1 = input("Enter name for Player 1 : ")
player1 = 'Johny Kob'
# player2 = input("Enter name for Player 2 : ")
player2 = 'Samoa Rob'

fight_start(player1,player2)
```
